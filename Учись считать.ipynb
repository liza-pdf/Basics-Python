```python
К Вашему вниманию предоставлен чат-бот, который обучает начальной арифметике. 
Чат-бот формирует случайный пример со случайным знаком операции (сложение, вычитание, деление, умножение). 
Пользователю необходимо ввести ответ на данный пример. 
Есть возможность накапливать очки за верные ответы (их можно тратить на подсказки). 
В конце программы чат-бот предоставляет общую статистику за игровую сессию.
```

```python
Метод isdigit() возвращает True, если все символы в строке являются цифрами, а если нет - возвращается False. 
В данной программе метод isdigit используется для проверки ввода типа данных пользователем.
Если тип данных не равно integer (целое число), то запрос ввода повторится.
Документация к данному методу: https://python-reference.readthedocs.io/en/latest/docs/str/isdigit.html

Метод upper() преобразует все символы нижнего регистра в строке в символы верхнего регистра и возвращает их. 
В данной программе метод upper используется для удобства пользователя, позволяя вводить команды в нижнем регистре.
Документация к данному методу: https://python-reference.readthedocs.io/en/latest/docs/str/upper.html
```

```python
import random as rd  # импортируем встроенную библиотеку случайных чисел

lowRange = 10  # нижний предел возможных ответов
highRange = 100  # верхний предел возможных ответов
allOperation = ['+', '-', '*', '/']  # знаки операций
x = 0  # первое число
y = 0  # второе число
z = 0  # решение
operation = ''  # один из знаков операций будет храниться
response = ''  # ответ пользователя
allScore = 0  # всего очков
allTask = 0  # всего решенных примеров
allCorrectSolution = 0  # всего правильных решенных примеров

print('*' * 27 + 'УЧИСЬ СЧИТАТЬ' + '*' * 27)
print(f'''Компьютер загадал Вам пример, в диапазоне результата от {lowRange} до {highRange}.
Сможете ли Вы правильно ответить?
Для подсказки введите HELP или ? (стоимость 15 очков)
Для выхода из программы введите STOP.''')
print('*' * 67)

# пока ПРАВДА (следовательно цикл бесконечный и единственная возможность выхода из него - оператор break)
while True:
    # переменная будет хранить случайный операнд из доступных в списке
    operation = rd.choice(allOperation)

    # если выпал знак СЛОЖЕНИЕ
    if (operation == '+'):
        # метод 'наоборот': узнаем в начале сумму из диапазона
        z = rd.randint(lowRange, highRange)
        # а затем первое слагаемое из диапазона (-1 нужен, чтоб Y не стал 0)
        x = rd.randint(lowRange, z) - 1
        # решение
        y = z - x

        # X чаще больше, чем Y, поэтому меняем слогаемые местами с помощью рандома
        if (rd.randint(0, 1) == 0):
            print(f'Пример: {x} + {y} = ?')
        else:
            print(f'Пример: {y} + {x} = ?')

    # если выпал знак ВЫЧИТАНИЯ
    if (operation == '-'):
        # уменьшаемое число из диапазона
        x = rd.randint(lowRange, highRange)
        # вычитаемое число из диапазона
        y = rd.randint(0, x - lowRange)
        # решение
        z = x - y

        print(f'Пример: {x} - {y} = ?')

    # если выпал знак УМНОЖЕНИЯ
    if (operation == '*'):
        # находим 1-ый множитель: (rd.randint(lowRange // 5, highRange // 10)) - для вариативности примеров;
        # + 1 - чтобы не получить x = 0
        x = rd.randint(1, highRange) // (rd.randint(lowRange // 5, highRange // 10)) + 1
        # находим 2-ый множитель: (x...) - исключаем появление 0;
        # (...+ lowRange...) - исключаем произведение меньше 10;
        # (...-1) - делаем возможно появления произведения = 10;
        # // x - чтобы не выйти за диапазон
        y = rd.randint((x + lowRange - 1), highRange) // x
        # решение
        z = x * y

        print(f'Пример: {x} * {y} = ?')

    # если выпал знак УМНОЖЕНИЯ
    if (operation == '/'):  # ДЕЛЕНИЕ
        # находим делимое: (rd.randint(lowRange // 2, highRange // 10)) - для вариативности примеров;
        # + 1 - чтобы не получить x = 0
        x = rd.randint(1, highRange) // (rd.randint(lowRange // 2, highRange // 10)) + 1
        # находим делитель: (x...) - исключаем появление 0;
        # // x - чтобы не выйти за диапазон
        y = rd.randint(x, highRange) // x
        # решение
        x = x * y
        z = x // y

        print(f'Пример: {x} / {y} = ?')

    # ПРОВЕРКА НА ВВОДИМЫЕ ДАННЫЕ
    # пока вводимые данные НЕ число, stop (или его вариации) или help (или его вариации)
    while (not response.isdigit()) \
            and (response.upper() != 'STOP') \
            and (response.upper() != 'ЫЕЩЗ') \
            and (response.upper() != 'S') \
            and (response.upper() != 'H') \
            and (response.upper() != 'РУДЗ') \
            and (response.upper() != 'HELP') \
            and (response != '?'):
        response = input(f'Ваш ответ: ')

        # если это не один из доступных вариантов ввода, то повторяем ввод (исключение ошибки ввода)
        if (not response.isdigit()) \
                and (response.upper() != 'STOP') \
                and (response.upper() != 'ЫЕЩЗ') \
                and (response.upper() != 'S') \
                and (response.upper() != 'H') \
                and (response.upper() != 'РУДЗ') \
                and (response.upper() != 'HELP') \
                and (response != '?'):
            print('Ошибка ввода: введите целое число, STOP или HELP!')

        # если это варианты вызова помощи, то проверяем кол-во очков
        elif (response.upper() == 'H') \
                or (response.upper() == 'РУДЗ') \
                or (response.upper() == 'HELP') \
                or (response == '?'):
            # если очков недостаточно, то подсказка недоступна
            if (allScore < 15):
                print('Недостаточно очков для подсказки!')
                response = input(f'Ваш ответ: ')
            # если ответ имеет одну цифру
            elif (z < 10):
                print('Ответ однозначный.')
                response = input(f'Ваш ответ: ')
                allScore -= 15
            # иначе показываем крайнию цифру ответа
            else:
                print(f'Крайняя цифра числа = {z % 10}')  # узнаем крайнию цифру числа
                response = input(f'Ваш ответ: ')
                allScore -= 15

    # выход из цикла
    if (response.upper() == 'STOP') \
            or (response.upper() == 'ЫЕЩЗ') \
            or (response.upper() == 'S'):
        break

    # изменяем тип данных переменной со строки в число (для дальнейшего сравнения с переменной z)
    response = int(response)

    # если ответ пользователя и правильный ответ одинаковы, то + к статистике
    if (z == response):
        allScore += 10  # добавляем очки
        allTask += 1  # счетчик кол-ва решенных примеров
        allCorrectSolution += 1  # счетчик правильно решенных примеров

        print('Верно!')
        print(f'Очков: {allScore}')
        print(f'Решенных задач: {allTask}')
        print(f'Правильно решенных задач: {allCorrectSolution}')
        print('-' * 67)

        # сброс переменной
        response = ''

    # если ответ пользователя и правильный ответ разные, то - к статистике
    else:
        allTask += 1  # всего решенных примеров

        print(f'Не верно! Правильный ответ: {z}')
        print(f'Очков: {allScore}')
        print(f'Решенных задач: {allTask}')
        print(f'Правильно решенных задач: {allCorrectSolution}')
        print('-' * 67)

        # сброс переменной
        response = ''

# после выхода из цикла (игры), предоставляется статистика игры
print('-' * 67)
print(f'''Ваша статистика за игру:
Всего решено примеров: {allTask}.
Верных ответов: {allCorrectSolution}.
Допущено ошибок: {allTask - allCorrectSolution}''')
if (allTask > 0):
    print(f'Процент верных ответов: {allCorrectSolution * 100 / allTask:.2f}%')
    print(f'Процент неверных ответов: {100 - (allCorrectSolution * 100 / allTask):.2f}%')
else:
    print('Процент верных ответов: 0')
    print('Процент неверных ответов: 0')
print('Спасибо Вам за игру!')
```
